```python
class StudentManager:
    def __init__(self):
        # List to store student dictionaries
        self.students = []

    def calculate_grade(self, marks):
        if marks >= 90: return 'A'
        elif marks >= 80: return 'B'
        elif marks >= 70: return 'C'
        else: return 'F'

    def add_student(self):
        print("\n--- Add New Student ---")
        try:
            roll = input("Enter Roll Number: ")
            # Check if roll number already exists
            if any(s['Roll Number'] == roll for s in self.students):
                print("Error: Roll Number already exists!")
                return

            name = input("Enter Name: ")
            dept = input("Enter Department: ")
            marks = float(input("Enter Marks: "))

            student = {
                "Roll Number": roll,
                "Name": name,
                "Department": dept,
                "Marks": marks,
                "Grade": self.calculate_grade(marks)
            }
            self.students.append(student)
            print(f"Student {name} added successfully!")
        except ValueError:
            print("Invalid input! Marks must be a number.")

    def view_students(self):
        print("\n--- All Student Records ---")
        if not self.students:
            print("No students found.")
            return

        for s in self.students:
            print(f"Roll: {s['Roll Number']} | Name: {s['Name']} | Dept: {s['Department']} | Marks: {s['Marks']} | Grade: {s['Grade']}")

    def search_student(self):
        roll = input("\nEnter Roll Number to search: ")
        for s in self.students:
            if s['Roll Number'] == roll:
                print(f"Found: {s}")
                return
        print("Error: Student not found.")

    def update_marks(self):
        roll = input("\nEnter Roll Number to update marks: ")
        for s in self.students:
            if s['Roll Number'] == roll:
                try:
                    new_marks = float(input(f"Enter new marks for {s['Name']}: "))
                    s['Marks'] = new_marks
                    s['Grade'] = self.calculate_grade(new_marks)
                    print("Marks and Grade updated successfully!")
                    return
                except ValueError:
                    print("Invalid input! Update failed.")
                    return
        print("Error: Student not found.")

    def delete_student(self):
        roll = input("\nEnter Roll Number to delete: ")
        for i, s in enumerate(self.students):
            if s['Roll Number'] == roll:
                del self.students[i]
                print(f"Student with Roll Number {roll} deleted.")
                return
        print("Error: Roll Number does not exist.")

# Main Program Loop
def main():
    sms = StudentManager()

    while True:
        print("\n--- STUDENT MANAGEMENT SYSTEM ---")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Update Student Marks")
        print("5. Delete Student")
        print("6. Exit")

        choice = input("\nSelect an option (1-6): ")

        if choice == '1':
            sms.add_student()
        elif choice == '2':
            sms.view_students()
        elif choice == '3':
            sms.search_student()
        elif choice == '4':
            sms.update_marks()
        elif choice == '5':
            sms.delete_student()
        elif choice == '6':
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```
