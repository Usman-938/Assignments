```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import sys
import pandas as pd
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, utils
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split

# ==========================================
# PART 1: DATASET EXPLORATION
# ==========================================
# Load CIFAR-10 dataset #[cite: 3, 8]
(x_train_full, y_train_full), (x_test, y_test) = datasets.cifar10.load_data()
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'] # [cite: 6]

# Report dataset shape and number of classes #[cite: 4, 9]
print(f"Dataset Shape: {x_train_full.shape}") # 60,000 images, 32x32x3 #[cite: 4]
print(f"Number of Classes: {len(np.unique(y_train_full))}") # 10 classes #[cite: 6]

# Display 10 sample images (one from each class) #[cite: 9]
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    idx = np.where(y_train_full == i)[0][0]
    plt.imshow(x_train_full[idx])
    plt.title(class_names[i])
    plt.axis('off')
plt.tight_layout()
plt.show()

# Analyze pixel value distribution #[cite: 9]
plt.figure(figsize=(6, 4))
plt.hist(x_train_full.ravel(), bins=50, color='blue', alpha=0.7)
plt.title("Pixel Value Distribution")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()

# ==========================================
# PART 2: DATA PREPROCESSING
# ==========================================
# Report memory usage before preprocessing #[cite: 13]
mem_before = x_train_full.nbytes / (1024**2)

# Normalize pixel values (0–1 range) #[cite: 11]
x_train_full = x_train_full.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convert labels to categorical format (one-hot encoding) #[cite: 12]
y_train_cat = utils.to_categorical(y_train_full, 10)
y_test_cat = utils.to_categorical(y_test, 10)

# Split validation set from training data #[cite: 12]
x_train, x_val, y_train, y_val = train_test_split(x_train_full, y_train_cat, test_size=0.2, random_state=42)

# Report memory usage after preprocessing #[cite: 13]
mem_after = x_train_full.nbytes / (1024**2)
print(f"Memory Usage Before: {mem_before:.2f} MB")
print(f"Memory Usage After: {mem_after:.2f} MB")

# ==========================================
# PART 3: MODEL 1 – MULTI-LAYER PERCEPTRON (MLP)
# ==========================================
# Build MLP with 3 Dense layers and ReLU #[cite: 15, 16]
mlp = models.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)), # 3072 features #[cite: 15]
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

mlp.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
mlp.summary() # Requirement: Model Summary #[cite: 34]

print("\nTraining MLP...")
start_time = time.time()
history_mlp = mlp.fit(x_train, y_train, epochs=20, validation_data=(x_val, y_val), batch_size=64, verbose=1) # [cite: 16]
mlp_time = time.time() - start_time
mlp_eval = mlp.evaluate(x_test, y_test_cat, verbose=0) # [cite: 18]

# ==========================================
# PART 4: MODEL 2 – CONVOLUTIONAL NEURAL NETWORK (CNN)
# ==========================================
# Build CNN with 3 conv layers, MaxPooling, and Dropout #[cite: 20, 21, 22]
cnn = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn.summary() # Requirement: Model Summary #[cite: 34]

print("\nTraining CNN...")
start_time = time.time()
history_cnn = cnn.fit(x_train, y_train, epochs=20, validation_data=(x_val, y_val), batch_size=64, verbose=1) # [cite: 23]
cnn_time = time.time() - start_time
cnn_eval = cnn.evaluate(x_test, y_test_cat, verbose=0) # [cite: 25]

# ==========================================
# PART 5: DATA AUGMENTATION EXPERIMENT
# ==========================================
# Apply rotation, flipping, and zoom #[cite: 27]
data_augmentation = models.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])

# Retrain CNN with augmented data #[cite: 27]
cnn_aug = models.Sequential([
    layers.Input(shape=(32, 32, 3)),
    data_augmentation,
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

cnn_aug.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("\nTraining CNN with Data Augmentation...")
history_aug = cnn_aug.fit(x_train, y_train, epochs=20, validation_data=(x_val, y_val), batch_size=64, verbose=1)
aug_eval = cnn_aug.evaluate(x_test, y_test_cat, verbose=0)

# ==========================================
# PART 6: MODEL COMPARISON & SUBMISSION
# ==========================================

# Training vs Validation Curves #[cite: 17, 24, 34]
def plot_history(history, title):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{title} Accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} Loss')
    plt.legend()
    plt.show()

plot_history(history_mlp, "MLP")
plot_history(history_cnn, "CNN")

# Comparison Table #[cite: 30, 35]
comparison_df = pd.DataFrame({
    "Metric": ["Accuracy", "Loss", "Training Time (s)", "Parameters"],
    "MLP": [mlp_eval[1], mlp_eval[0], mlp_time, mlp.count_params()],
    "CNN": [cnn_eval[1], cnn_eval[0], cnn_time, cnn.count_params()],
    "CNN (Augmented)": [aug_eval[1], aug_eval[0], "N/A", cnn_aug.count_params()]
})
print("\n--- Model Comparison Table ---")
print(comparison_df)

# Confusion Matrix (Standard CNN) #[cite: 30, 34]
y_pred = np.argmax(cnn.predict(x_test), axis=1)
cm = confusion_matrix(np.argmax(y_test_cat, axis=1), y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('CNN Confusion Matrix')
plt.show()

# Display 10 misclassified images (Standard CNN) #[cite: 31, 35]
misclassified_idx = np.where(y_pred != np.argmax(y_test_cat, axis=1))[0]
plt.figure(figsize=(12, 6))
for i, idx in enumerate(misclassified_idx[:10]):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_test[idx])
    plt.title(f"Act: {class_names[np.argmax(y_test_cat[idx])]} \nPred: {class_names[y_pred[idx]]}", fontsize=8)
    plt.axis('off')
plt.suptitle("Misclassified Image Samples")
plt.tight_layout()
plt.show()
```
